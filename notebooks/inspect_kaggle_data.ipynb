# Inspect Kaggle Augmented Data

This notebook loads the `data/training/kaggle_legal_augmented.parquet` file and displays sample rows to verify the data augmentation (Summary + Explanation).

```python
import pandas as pd
import os

pd.set_option('display.max_colwidth', None)

file_path = "../data/training/kaggle_legal_augmented.parquet"

if os.path.exists(file_path):
    df = pd.read_parquet(file_path)
    print(f"Loaded dataset with {len(df)} rows.")
    print("Columns:", df.columns.tolist())
else:
    print(f"File not found at {file_path}")
```

## 1. Inspect Random Samples

```python
if 'df' in locals():
    sample = df.sample(3)
    for idx, row in sample.iterrows():
        print(f"\n--- Sample {idx} ---")
        print(f"[INPUT (Instruction + Text snippet)]:\n{row['input'][:500]}...\n")
        print(f"[OUTPUT (Target: Summary + Explanation)]:\n{row['output']}\n")
        print("-"*80)
```

## 2. Check Split Distribution

```python
if 'df' in locals() and 'split' in df.columns:
    print(df['split'].value_counts())
```
